# **Data and Libraries**

```python
import matplotlib.pyplot as plt

time = [0, 30, 60, 90, 120, 150, 180, 210, 240, 270, 300]
dark_assay = [0.28, 0.27, 0.26, 0.27, 0.27, 0.27, 0.26, 0.25, 0.25, 0.25, 0.24]
light_assay = [0.49, 0.49, 0.47, 0.46, 0.45, 0.45, 0.44, 0.43, 0.42, 0.42, 0.41]
green_assay = [0.38, 0.38, 0.37, 0.36, 0.38, 0.35, 0.38, 0.34, 0.36, 0.36, 0.35]
```

# **The Figure**

```python
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np

plt.figure(figsize=(8, 6))
plt.plot(time, dark_assay, marker='o', linestyle='-', label='Dark Assay', color='blue')
plt.plot(time, light_assay, marker='o', linestyle='-', label='Light Assay', color='orange')
plt.plot(time, green_assay, marker='o', linestyle='-', label='Green Assay', color='green')

for assay, color, label in zip([dark_assay, light_assay, green_assay], ['blue', 'orange', 'green'],
                               ['Dark Assay', 'Light Assay', 'Green Assay']):
    slope, intercept, r_value, _, _ = linregress(time, assay)
    time_array = np.array(time)
    plt.plot(time, slope * time_array + intercept, linestyle='--', color=color,
             label=f'{label} Regression (RÂ²={r_value**2:.2f})')

    equation = f'y = {slope:.4f}x + {intercept:.4f}'
    plt.text(time[-1], (slope * time_array[-1] + intercept) - 0.01, equation,
             color=color, ha='right', va='top')



plt.grid(True)
plt.xlabel('Time (s)')
plt.ylabel('Absorbance')
plt.title('Absorbance of Light at 620 nm in Varying Conditions')
plt.legend()
plt.tight_layout()
plt.show()
```
